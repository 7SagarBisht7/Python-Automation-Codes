```python
import threading
import time
def task():
  for i in range(3):
    print("Thread running: ",i)
    time.sleep(1)

t1 = threading.Thread(target=task)
t1.start()
t1.join()
print("Main Thread finished")
```

```python
import threading
class MyThread(threading.Thread):
  def run(self):
      print("Thread running")

t1 = MyThread()
t1.start()
t1.join()
print("Main Thread finished")
```

```python
import threading
import time

def task():
  print("Thread started")
  time.sleep(3)
  print("Thread finished")

t1 = threading.Thread(target=task)

print("Before start  - Is thread alive?",t1.is_alive())
t1.start()
time.sleep(1)
print("After start  - Is thread alive?",t1.is_alive())
t1.join()
print("After join  - Is thread alive?",t1.is_alive())
```

```python
import threading
import time

def task():
  print("Thread started")
  time.sleep(3)
  print("Thread finished")

t = threading.Thread(target=task)
t.setDaemon(True)
t.start()

print("Main program completed!")
```
