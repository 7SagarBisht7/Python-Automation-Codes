```python
import scrapy
import nest_asyncio
from scrapy.crawler import CrawlerProcess
nest_asyncio.apply()
```

```python
class ServerSidePaginationSpider(scrapy.Spider):
  name = "server_side_pagination"
  start_urls = ["https://quotes.toscrape.com/"]
  def parse(self, response):
    print(f"Scraped page: {response.url}")
    for quote in response.css("div.quote"):
      text = quote.css("span.text::text").get()
      author = quote.css("small.author::text").get()
      print("Quotes:",text)
      print("Author:",author)
      print("-"*50)
    next_page = response.css("li.next a::attr(href)").get()
    if next_page:
      next_page_url = response.urljoin(next_page)
      print(f"Requesting Next Page:{next_page_url}")
      yield scrapy.Request(next_page_url, callback=self.parse)

process = CrawlerProcess(settings={"LOG_LEVEL":"ERROR"})
process.crawl(ServerSidePaginationSpider)
process.start()
```

```python
class ClientSidePaginationSpider(scrapy.Spider):
  name = "client_side_pagination"
  def start_requests(self):
    for page in range(1,4):
      api_url = f"https://quotes.toscrape.com/api/quotes?page={page}"
      print(f"Requesting API URL: {api_url}")
      yield scrapy.Request(api_url,callback = self.parse)
  def parse(self, response):
    data = response.json()
    print("Scraped API Page")
    for quote in data["quotes"]:
      print("Quotes:",quote["text"])
      print("Author:",quote["author"]["name"])
      print("-"*50)

# Create a new CrawlerProcess instance for each spider run to avoid ReactorNotRestartable error
process = CrawlerProcess(settings={"LOG_LEVEL":"ERROR"})
process.crawl(ClientSidePaginationSpider)
process.start()
```
