```python
import threading
count = 0
lock = threading.Lock()
def increment():
  global count
  for _ in range(1000000):
    lock.acquire()
    count += 1
    lock.release()

t1 = threading.Thread(target=increment)
t2 = threading.Thread(target=increment)
t1.start()
t2.start()
t1.join()
t2.join()
print("Final count:",count)
```

```python
import threading
rlock = threading.RLock()
def outer_function():
    print("Outer function acquring lock")
    rlock.acquire()
    print("Outer function acqured lock")
    inner_function()
    rlock.release()
    print("Outer function released lock")

def inner_function():
    print("Inner function acquring lock")
    rlock.acquire()
    print("Inner function acqured lock")
    rlock.release()
    print("Inner function released lock")

t = threading.Thread(target=outer_function)
t.start()
t.join()
```
