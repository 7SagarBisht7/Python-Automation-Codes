```python
import pandas as pd
import datetime
from datetime import date, time, datetime, timedelta
```

```python
df = pd.read_csv('/content/login_activity.csv')
display(df.head())
```

```python
df['login_time'] = df['login_time'].apply(lambda x: datetime.strptime(x, '%d-%m-%Y %H:%M:%S'))
df['logout_time'] = df['logout_time'].apply(lambda x: datetime.strptime(x, '%d-%m-%Y %H:%M:%S'))

display(df.head())
```

```python
df['session_duration'] = df['logout_time'] - df['login_time']
df['session_duration_minutes'] = df['session_duration'].dt.total_seconds() / 60

display(df.head())
```

```python
df['login_date'] = df['login_time'].dt.date
logins_per_day = df.groupby('login_date').size().reset_index(name='number_of_logins')
logins_per_day = logins_per_day.sort_values(by='login_date', ascending=True)
display(logins_per_day)
```
