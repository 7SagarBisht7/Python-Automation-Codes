```python
import socket, threading, time
def server():
  s = socket.socket(socket.AF_INET,socket.SOCK_STREAM)
  s.bind(('localhost',5000))
  s.listen(1)
  print("Sever listening....")
  conn, addr = s.accept()
  print("Connected by ", addr)
  data = conn.recv(1024)
  print("Server receieved:", data.decode())
  conn.send(b"ACK")
  conn.close()
  s.close

threading.Thread(target=server, daemon = True).start()
time.sleep(1)
```

```python
client = socket.socket(socket.AF_INET,socket.SOCK_STREAM)
client.settimeout(5)
client.connect(('localhost',5000))
client.send(b"Hello Server")

print("Client receieved:",client.recv(1024).decode())
client.close()
```

```python
import socket, threading, time

def udp_server():
    s = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
    s.bind(('localhost', 5002))
    print("UDP server running...")
    data, addr = s.recvfrom(1024)
    print("UDP server received:", data.decode())
    s.sendto(b"ACK from UDP server", addr)
    s.close()

threading.Thread(target=udp_server, daemon=True).start()
time.sleep(1)
c = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
c.sendto(b"Hello UDP server", ('localhost', 5002))
print("UDP client received:", c.recvfrom(1024)[0].decode())
c.close()
```
