```python
import threading
import time
N = 5
forks = [threading.Semaphore(1) for _ in range(N)]
def philosopher(i):
    left = forks[i]
    right = forks[(i+1)%N]
    for _ in range(2):
          print(f"Philosopher {i} is thinking")
          time.sleep(1)
          first,second = (left,right) if i%2 == 0 else (right,left)
          with first:
            with second:
              print(f"Philosopher {i} is eating")
              time.sleep(2)
          print(f"Philosopher {i} finished eating\n")
threads = []
for i in range(N):
  t = threading.Thread(target = philosopher, args = (i,))
  threads.append(t)
  t.start()
for t in threads:
  t.join()
print("Dinner is over")
```

```python
Students(threads) must wait until the exam starts. Once the exam begins, all students can attempt it simultaneously.
Requirements: Use condition. Students wait for an "exam started" signal
Examiner thread signals start
```

```python
import threading
import time

# Create a Condition object
exam_started_condition = threading.Condition()

# Flag to indicate if the exam has started
exam_started = False

def student(student_id):
    print(f"Student {student_id}: Waiting for the exam to start...")
    with exam_started_condition:
        while not exam_started:
            exam_started_condition.wait() # Wait until exam_started is True
        print(f"Student {student_id}: Exam has started! Attempting the exam...")
        time.sleep(2) # Simulate taking the exam
        print(f"Student {student_id}: Finished the exam.")

def examiner():
    global exam_started
    print("Examiner: Preparing the exam materials...")
    time.sleep(5) # Simulate examiner preparation time
    print("Examiner: Signalling all students to start the exam now!")
    with exam_started_condition:
        exam_started = True
        exam_started_condition.notify_all() # Wake up all waiting student threads
    print("Examiner: Exam has officially begun.")

# Create student threads
num_students = 5
student_threads = []
for i in range(num_students):
    t = threading.Thread(target=student, args=(i + 1,))
    student_threads.append(t)
    t.start()

# Create and start the examiner thread
examiner_thread = threading.Thread(target=examiner)
examiner_thread.start()

# Wait for all threads to complete
examiner_thread.join()
for t in student_threads:
    t.join()

print("All students have finished their exams. Session concluded.")
```

```python
import threading
import time
condition = threading.Condition()
exam_started = False
def student(student_id):
  global exam_started
  with condition:
    print(f"Student {student_id}: Waiting for the exam to start...")
    while not exam_started:
      condition.wait()
  print(f"STUDENT {student_id}: Exam has started! Attempting the exam...")
  time.sleep(2)
  print(f"STUDENT {student_id}: Finished the exam.")

def examiner():
  global exam_started
  time.sleep(3)
  with condition:
    print("Examiner: Exam is starting now!...")
    exam_started = True
    condition.notify_all()

students = []
for i in range(1,6):
  t = threading.Thread(target = student, args = (i,))
  students.append(t)
  t.start()

examiner_thread = threading.Thread(target = examiner)
examiner_thread.start()
for t in students:
  t.join()

examiner_thread.join()
print("\n All students have completed the exam")
```
