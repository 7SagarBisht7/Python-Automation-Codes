```python
import threading
import time
sem = threading.Semaphore(2)
def task(name):
  print(f"{name} waiting to enter")
  sem.acquire()
  print(f"{name} entered critical section")
  time.sleep(2)
  print(f"{name} leaving critical section")
  sem.release()
threads = []
for i in range(5):
  t = threading.Thread(target=task,args=(f"thread{i+1}",))
  threads.append(t)
  t.start()
for t in threads:
  t.join()
print("All threads finished")
```

```python
import threading
import time
buffer = []
MAX_SIZE = 5
condition = threading.Condition()
def producer():
  for i in range(10):
    with condition:
      while len(buffer) == MAX_SIZE:
        condition.wait()
      buffer.append(i)
      print(f"Produced:{i}")
      condition.notify()
    time.sleep(1)

def consumer():
  for _ in range(10):
    with condition:
      while not buffer:
        condition.wait()
      item = buffer.pop(0)
      print(f"Consumed: {item}")
      condition.notify()
    time.sleep(1)

t1 = threading.Thread(target=producer)
t2 = threading.Thread(target=consumer)
t1.start()
t2.start()
t1.join()
t2.join()
print("Producer and Consumer finished")
```
