```python
from datetime import datetime

def convert_two_date_inputs():
    date_objects = []
    for i in range(1, 3):
        date_str = input(f"Please enter date {i} in DD-MM-YYYY format (e.g., 25-12-2023): ")
        try:
            date_obj = datetime.strptime(date_str, '%d-%m-%Y').date()
            print(f"Successfully converted date {i}! The date object is: {date_obj}")
            print(f"Type of the object: {type(date_obj)}")
            date_objects.append(date_obj)
        except ValueError:
            print(f"Invalid date format for date {i}: '{date_str}'. Skipping this date and assigning None.")
            date_objects.append(None) # Append None to maintain list length for unpacking
    return date_objects

date1, date2 = convert_two_date_inputs()
print(f"\nFirst date: {date1}, Second date: {date2}")
```

```python
if date1 < date2:
    print(f"Date 1 ({date1}) is earlier than Date 2 ({date2}).")
elif date1 > date2:
    print(f"Date 1 ({date1}) is later than Date 2 ({date2}).")
else:
    print(f"Date 1 ({date1}) and Date 2 ({date2}) are the same.")
```

```python
time_diff = date2-date1
days_diff = abs(time_diff.days)
print(f"The number of days between {date1} and {date2} is: {days_diff} days.")
```

```python
from datetime import datetime

def calculate_age():
    while True:
        dob_str = input("Please enter your date of birth (DOB) in DD-MM-YYYY format (e.g., 01-01-1990): ")
        try:
            dob = datetime.strptime(dob_str, '%d-%m-%Y').date()
            break
        except ValueError:
            print("Invalid date format. Please try again using DD-MM-YYYY.")

    today = datetime.today().date()
    age = today.year - dob.year - ((today.month, today.day) < (dob.month, dob.day))
    print(f"\nDate of Birth : {dob}")
    print(f"Today's Date: {today}")
    print(f"The person's age is: {age} years.")

calculate_age()
```

```python
from os.path import isfile
import os
from datetime import datetime, timedelta
path = input("Enter the path of the directory: ")
x_days = int(input("Enter the number of days: "))
current_time = datetime.now()
threshold_time = current_time - timedelta(days=x_days)
print(f"\nFiles older than {x_days} days:\n")
try:
  for file in os.listdir(path):
    file_path = os.path.join(path,file)
    if os.path.isfile(file_path):
      modified_time = datetime.fromtimestamp(os.path.getmtime(file_path))
      if modified_time < threshold_time:
        print(f"{file} --> Last Modified: {modified_time}")
except FileNotFoundError:
  print("Invalid path. Please enter a valid path.")
except PermissionError:
  print("Permission denied. Please check the path permissions.")
```
