```python
import socket
import threading
import time
SERVER_IP = "localhost"
SERVER_PORT = 7000
BUFFER_SIZE = 1024
def udp_chat_server():
  server = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
  server.bind((SERVER_IP, SERVER_PORT))
  print("UDP Chat Server started...\n")
  clients = set()
  while True:
    message, addr = server.recvfrom(BUFFER_SIZE)
    if addr not in clients:
      clients.add(addr)
      print(f"[SERVER] New client joined: {addr}")
    decoded_mesg = message.decode()
    print(f"[SERVER] Received message from {addr}: {decoded_mesg}")

    for client in clients:
      if client != addr:
        server.sendto(message, client)

def udp_chat_client(name,messages):
  client = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
  def receive():
    while True:
      try:
        msg, _ = client.recvfrom(BUFFER_SIZE)
        print(f"[{name} RECEIVED]:{msg.decode()}")
      except:
        break

  threading.Thread(target=receive, daemon = True).start()
  for msg in messages:
    full_msg = f"{name}:{msg}"
    client.sendto(full_msg.encode(), (SERVER_IP, SERVER_PORT))
    time.sleep(2)

  time.sleep(2)
  client.close()
```

```python
if __name__ == "__main__":
  threading.Thread(target=udp_chat_server, daemon=True).start()
  time.sleep(1)
  client1_msgs = ["Hello","How are you?","This is sample"]
  client2_msgs = ["Hi","I'm fine","Fast but unreal"]
  client3_msgs = ["Hey all","UDP chat demo","Byeee"]

  threading.Thread(target=udp_chat_client, args=("Client1",client1_msgs)).start()
  threading.Thread(target=udp_chat_client, args=("Client2",client2_msgs)).start()
  threading.Thread(target=udp_chat_client, args=("Client3",client3_msgs)).start()



```
