```python
import datetime
```

```python
from datetime import date, time, datetime, timedelta
```

```python
now = datetime.now()
print(now)
```

```python
today = date.today()
print(today)
print(today.year)
print(today.month)
print(today.day)
```

```python
t = time(10, 15, 30)
print(t)
print(t.hour)
print(t.minute)
print(t.second)
```

```python
t = datetime.now()
print(t.time())
print(t.hour)
print(t.minute)
print(t.second)
```

```python
now = datetime.now()
print(now)
```

```python
print(now.year, now.month, now.day)
print(now.hour, now.minute, now.second)
```

```python
#merge date and time
datetime.combine(today, now.time())
```

```python
formatted = now.strftime("%d-%m-%Y %H:%M:%S")
print(formatted)
```

```python
dt = datetime.strptime("22-01-2026", "%d-%m-%Y")
print(dt)
```

```python
#timedelta
from datetime import timedelta

future = now + timedelta(days=7)
print(future)
past = now - timedelta(hours =3)
print(past)
```
